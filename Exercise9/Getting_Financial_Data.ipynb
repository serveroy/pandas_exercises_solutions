### Step 5. Add a new column "stock" to the dataframe and add the ticker symbol

```python
df_apple["stock"] = "AAPL"
df_apple
```

### Step 6. Repeat the two previous steps for a few other stocks, always creating a new dataframe: Tesla, IBM and Microsoft. (Ticker symbols TSLA, IBM and MSFT.)

```python
# Tesla

df_tesla = yf.download("TSLA", start=start_date, end=end_date)
df_tesla.columns = [col[0] for col in df_tesla.columns]
df_tesla["stock"] = "TSLA"

# IBM

df_ibm = yf.download("IBM", start=start_date, end=end_date)
df_ibm.columns = [col[0] for col in df_ibm.columns]
df_ibm["stock"] = "IBM"

# Microsoft

df_microsoft = yf.download("MSFT", start=start_date, end=end_date)
df_microsoft.columns = [col[0] for col in df_microsoft.columns]
df_microsoft["stock"] = "MSFT"
```

### Step 7. Combine the four separate dataFrames into one combined dataFrame df that holds the information for all four stocks

```python
df = pd.concat([df_apple, df_tesla, df_ibm, df_microsoft])
df
```

### Step 8. Shift the stock column into the index (making it a multi-level index consisting of the ticker symbol and the date).

```python
df = df.set_index([df.index, df["stock"]])
del df["stock"]
```

### Step 7. Create a dataFrame called vol, with the volume values.

```python
vol = df[["Volume"]]
vol
```

### Step 8. Aggregate the data of volume to weekly.
Hint: Be careful to not sum data from the same week of 2015 and other years.

```python
grouping_by_week = df.groupby(["stock", pd.Grouper(level=0, freq="W")]).sum()
grouping_by_week
```

### Step 9. Find all the volume traded in the year of 2015

```python
grouping_by_year = grouping_by_week.groupby(["stock", pd.Grouper(level=1, freq="YE")]).sum()

# rows_list = []
# for i in range(len(grouping_by_year.index)):
#     if grouping_by_year.index[i][1].year == 2015:
#         rows_list.append(i)
# rows_list
# final_answer = grouping_by_year.iloc[rows_list, :]["Volume"].reset_index()

check_2015_list = grouping_by_year.index.map(lambda x: x[1].year == 2015)
final_answer = grouping_by_year.loc[check_2015_list == True]["Volume"].reset_index()

del final_answer["Date"]
final_answer["Year"] = "2015"

final_answer
```
