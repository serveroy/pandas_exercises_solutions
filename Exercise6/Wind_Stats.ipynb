### Step 3. Assign it to a variable called data and replace the first 3 columns by a proper datetime index.

```python
data = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data", sep=r"\s+")
# sep = r"\s+" works for this case of space seperator (sep = " " doesn't work)

data['Yr'] = ["20" + str(x) for x in data['Yr']]
data['date'] = pd.to_datetime(dict(year=data.Yr, month=data.Mo, day=data.Dy))

del data['Yr']
del data['Mo']
del data['Dy']

data = data.set_index('date')

data
```

### Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

```python
def decrease_year(dates_list, param):
    # Without for loop
    return dates_list - pd.DateOffset(years=param)
    
    # # With for loop (invalid)
    # return [pd.to_datetime(str(curr_date.year - param) + " " + str(curr_date.month) + " " + str(curr_date.day)) for curr_date in dates_list]

new_dates_list = decrease_year(data.index, 100)

# Is valid
print(new_dates_list[-1].year < 2026)

data
```

### Step 5. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].

```python
data = data.reset_index(drop=True)
data["date"] = new_dates_list
data = data.set_index("date")

data
# data = data.reindex(new_dates_list) didnt work
```

### Step 6. Compute how many values are missing for each location over the entire record.  
#### They should be ignored in all calculations below.

### Step 12. Downsample the record to a yearly frequency for each location.

```python
# Yearly_frequency = data.groupby(data.index.year)
Yearly_frequency = data.groupby(pd.Grouper(freq="YE"))
Yearly_frequency.count()
```

### Step 13. Downsample the record to a monthly frequency for each location.

```python
# Monthly_frequency = data.groupby(data.index.month)
Monthly_frequency = data.groupby(pd.Grouper(freq="ME"))
Monthly_frequency.count()
```

### Step 14. Downsample the record to a weekly frequency for each location.

```python
# Weekly_frequency = data.groupby(data.index.day // 7)
Weekly_frequency = data.groupby(pd.Grouper(freq="W"))
Weekly_frequency.count()
```
