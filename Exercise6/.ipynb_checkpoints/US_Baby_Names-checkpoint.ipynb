### Introduction:

We are going to use a subset of [US Baby Names](https://www.kaggle.com/kaggle/us-baby-names) from Kaggle.  
In the file it will be names from 2004 until 2014


### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv). 

### Step 3. Assign it to a variable called baby_names.

```python
baby_names = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv", sep = ",")
```

### Step 4. See the first 10 entries

```python
baby_names.head(10)
```

### Step 5. Delete the column 'Unnamed: 0' and 'Id'

```python
del baby_names["Id"]
del baby_names["Unnamed: 0"]
```

### Step 6. What year has the highest number of baby names in the dataset?

```python
# .count() counts everything in this Year, but unique() count unique objects (names in this case)
baby_names_by_year = baby_names.groupby(["Year"])[["Name"]].nunique()
list(baby_names_by_year.loc[baby_names_by_year["Name"] == max(baby_names_by_year["Name"])].index)[0]
```

### Step 7. Is there more male or female names in the dataset?

```python
baby_names_by_gender = baby_names.groupby(["Gender"])[["Gender"]].count()
list(baby_names_by_gender.loc[baby_names_by_gender["Gender"] == max(baby_names_by_gender["Gender"])].index)[0]
```

### Step 8. Group the dataset by name and assign to names

```python
names = baby_names.groupby(["Name"])
```

### Step 9. How many different names exist in the dataset?

```python
names.count().shape[0]
```

### Step 10. What is the name with most occurrences?

```python
names_frequencies = dict(names["Name"].count())
max(names_frequencies, key = names_frequencies.get)  # Find the key with the maximum val
names_frequencies['Riley']
```

### Step 11. How many different names have the least occurrences?

```python
min_occurrences = min(names_frequencies.values())
count = 0
for curr_name in names_frequencies.keys():
    count += (names_frequencies[curr_name] == min_occurrences)
print(count)
```

### Step 12. What is the median name occurrence?

```python
names["Name"].count().median()
```

### Step 13. What is the standard deviation of names?

```python
names["Name"].count().std()
```

### Step 14. Get a summary with the mean, min, max, std and quartiles.

```python
names["Name"].count().describe()
```
