### Step 1. Import the necessary libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv). 

### Step 3. Assign it to a variable called tips

```python
tips = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv", sep = ",")
```

### Step 4. Delete the Unnamed 0 column

```python
del tips["Unnamed: 0"]
```

### Step 5. Plot the total_bill column histogram

```python
tips["total_bill"].hist();
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
plt.scatter(tips["tip"], tips["total_bill"])
plt.xlabel("tip");
plt.ylabel("total_bill");
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(tips["tip"], tips["size"], tips["total_bill"], marker = '^')

ax.set_xlabel('tip')
ax.set_ylabel('size')
ax.set_zlabel('total_bill')

ax.set_box_aspect(aspect=None, zoom=0.84)
plt.show()
```

### Step 8. Present the relationship between days and total_bill value

```python
plt.scatter(tips["day"], tips["total_bill"])
plt.xlabel("days");
plt.ylabel("total_bill");

# histogram?
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
sex_list = [tips["sex"][i] == "Female" for i in range(len(tips["sex"]))]

plt.scatter(tips["tip"], tips["day"], c = sex_list)
plt.xlabel("tip");
plt.ylabel("day");
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
# Boxplot of total_bill per day, differentiated by time
sns.boxplot(x=tips["day"], y=tips["total_bill"], hue=tips["time"])
plt.title("Total Bill per Day by Time (Lunch vs Dinner)")
plt.show()
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
tips_dinner = tips.loc[tips["time"] == "Dinner"]
tips_lunch = tips.loc[tips["time"] == "Lunch"]

plt.subplot(2, 1, 1)
plt.hist(tips_dinner["tip"]);
plt.title("Histogram of tips in dinner")
plt.subplot(2, 1, 2)
plt.title("Histogram of tips in lunch")
plt.hist(tips_lunch["tip"]);
plt.tight_layout()  # Adds space between the plots
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
GenderList = ["Male", "Female"]
for i in range(len(GenderList)):
    tips_gender = tips.loc[tips["sex"] == GenderList[i]]
    smoker_list = [list(tips_gender["smoker"])[i] == "Yes" for i in range(len(tips_gender["smoker"]))]
    
    plt.subplot(2, 1, i + 1)
    plt.scatter(tips_gender["total_bill"], tips_gender["tip"], c=smoker_list);
    plt.title(f"Scatterplot of tips for {GenderList[i].lower() + 's'} ")
    plt.xlabel("total bill")
    plt.ylabel("tip")
    
plt.tight_layout()  # Adds space between the plots
```
