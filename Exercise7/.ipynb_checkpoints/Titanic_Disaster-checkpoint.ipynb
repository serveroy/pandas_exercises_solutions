### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv)

### Step 3. Assign it to a variable titanic 

```python
titanic = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/refs/heads/master/07_Visualization/Titanic_Disaster/train.csv", sep = ",")
```

### Step 4. Set PassengerId as the index 

```python
titanic.set_index("PassengerId")
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
sex_grouping = titanic.groupby(["Sex"])[["Sex"]].count()
fig, ax = plt.subplots()
ax.pie(list(sex_grouping["Sex"]), labels=list(sex_grouping.index), autopct='%1.1f%%');
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
gender_list = [titanic["Sex"][i] == "female" for i in range(len(titanic["Sex"]))]
plt.scatter(titanic["Age"], titanic["Fare"], c=gender_list);
```

### Step 7. How many people survived?

```python
titanic["Survived"].sum()
```

### Step 8. Create a histogram with the Fare payed

```python
plt.hist(titanic["Fare"]);
```

### BONUS: Create your own question and answer it.

```python
# Create a pie chart of the Pclass
Pclass_grouping = titanic.groupby(["Pclass"])[["Pclass"]].count()
fig, ax = plt.subplots()
ax.pie(list(Pclass_grouping["Pclass"]), labels=list(Pclass_grouping.index), autopct='%1.1f%%');
```
