### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv). 

### Step 3. Assign it to a variable called drinks.

```python
drinks = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv", sep = ',')
```

### Step 4. Which continent drinks more beer on average?

```python
# # Algorithmic solution without groupby

# beer_dict = {"AF": 0, "AS": 0, "EU": 0, "OC": 0, "SA": 0}
# for continent in list(beer_dict.keys()):
#     curr_beer_count = drinks.loc[drinks["continent"] == continent, "beer_servings"].sum()
#     curr_continent_count = drinks.loc[drinks["continent"] == continent, "continent"].shape[0]
#     beer_dict[continent] = curr_beer_count/curr_continent_count

# beer_avg_list = list(beer_dict.values())
# max_beer_index = beer_avg_list.index(max(beer_avg_list))
# max_continent = list(beer_dict.keys())[max_beer_index]

# print(max_continent)

# # With groupby, not optimal

# beer_servings_summary = drinks.groupby(["continent"])["beer_servings"].mean()
# continent_list = list(dict(beer_servings_summary).keys())
# beer_servings_summary.sort_values().tail(1)

beer_servings_summary = drinks.groupby(["continent"])["beer_servings"].mean()
beer_servings_summary.idxmax()
```

### Step 5. For each continent print the statistics for wine consumption.

```python
drinks.groupby(["continent"])["wine_servings"].mean()
```

### Step 6. Print the mean alcohol consumption per continent for every column

```python
drinks.groupby(["continent"])[drinks.columns[1:-1]].mean()
```

### Step 7. Print the median alcohol consumption per continent for every column

```python
drinks.groupby(["continent"])[drinks.columns[1:-1]].median()
```

### Step 8. Print the mean, min and max values for spirit consumption.
#### This time output a DataFrame

```python
mean_spirit = drinks.groupby(["continent"])["spirit_servings"].mean()
min_spirit = drinks.groupby(["continent"])["spirit_servings"].min()
max_spirit = drinks.groupby(["continent"])["spirit_servings"].max()

spirit_statistics_dict = {"continent": mean_spirit.index, "mean": mean_spirit, "min": min_spirit, "max": max_spirit}
spirit_statistics = pd.DataFrame(spirit_statistics_dict)
spirit_statistics.set_index("continent")
```
