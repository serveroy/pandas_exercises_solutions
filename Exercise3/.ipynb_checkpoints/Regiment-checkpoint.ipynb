### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Create the DataFrame with the following values:

```python
raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'], 
        'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'], 
        'name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze', 'Jacon', 'Ryaner', 'Sone', 'Sloan', 'Piger', 'Riani', 'Ali'], 
        'preTestScore': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
        'postTestScore': [25, 94, 57, 62, 70, 25, 94, 57, 62, 70, 62, 70]}
```

### Step 3. Assign it to a variable called regiment.
#### Don't forget to name each column

```python
regiment = pd.DataFrame(raw_data, columns=raw_data.keys())
```

### Step 4. What is the mean preTestScore from the regiment Nighthawks?  

```python
all_means_preTestScore = regiment.groupby(["regiment"])[["preTestScore"]].mean()
all_means_preTestScore.loc["Nighthawks"]
```

### Step 5. Present general statistics by company

```python
# regiment.groupby(["company"])[["preTestScore", "postTestScore"]].mean()
# Only mean? How can I present all statistics, by company?

regiment.groupby(["company"])[["name", "preTestScore", "postTestScore"]].agg(
    # name_count = ("name", "count"),
    
    preTestScore_min = ("preTestScore", "min"),
    preTestScore_max = ("preTestScore", "max"),
    preTestScore_mean = ("preTestScore", "mean"),
    preTestScore_median = ("preTestScore", "min"),

    postTestScore_min = ("postTestScore", "min"),
    postTestScore_max = ("postTestScore", "max"),
    postTestScore_mean = ("postTestScore", "mean"),
    postTestScore_median = ("postTestScore", "min")   
    )
```

### Step 6. What is the mean of each company's preTestScore?

```python
regiment.groupby(["company"])[["preTestScore"]].mean()
```

### Step 7. Present the mean preTestScores grouped by regiment and company

```python
regiment.groupby(["regiment", "company"])[["preTestScore"]].mean()
```

### Step 8. Present the mean preTestScores grouped by regiment and company without heirarchical indexing

```python
regiment.groupby(["regiment", "company"], sort=False)[["preTestScore"]].mean()
```

### Step 9. Group the entire dataframe by regiment and company

```python
final_grouping = regiment.groupby(["regiment", "company"])
```

### Step 10. What is the number of observations in each regiment and company

```python
final_grouping[["name"]].agg(count = ("name", "count"))
```

### Step 11. Iterate over a group and print the name and the whole data from the regiment

```python
regiment_list = regiment["regiment"].unique()
for curr_regiment in regiment_list:
    print(curr_regiment)
    print(regiment.loc[regiment["regiment"] == curr_regiment])
```
