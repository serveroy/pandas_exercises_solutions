# Housing Market

### Introduction:

This time we will create our own dataset with fictional numbers to describe a house market. As we are going to create random data don't try to reason of the numbers.

### Step 1. Import the necessary libraries

```python
import pandas as pd
import random
```

### Step 2. Create 3 differents Series, each of length 100, as follows: 
1. The first a random number from 1 to 4 
2. The second a random number from 1 to 3
3. The third a random number from 10,000 to 30,000

```python
first = pd.Series(random.choices(range(1, 5), k=100))
second = pd.Series(random.choices(range(1, 4), k=100))
third = pd.Series(random.choices(range(10000, 30001), k=100))

print(first)
print("----------------")
print(second)
print("----------------")
print(third)
```

### Step 3. Let's create a DataFrame by joinning the Series by column

```python
joining_all_column = pd.concat([first, second, third], axis=1)
joining_all_column.head(5)
```

### Step 4. Change the name of the columns to bedrs, bathrs, price_sqr_meter

```python
joining_all_column = joining_all_column.set_axis(['bedrs','bathrs', 'price_sqr_meter'], axis=1)
joining_all_column.head(5)
```

### Step 5. Create a one column DataFrame with the values of the 3 Series and assign it to 'bigcolumn'

```python
bigcolumn = pd.concat([first, second, third], axis=0)
bigcolumn
```

### Step 6. Oops, it seems it is going only until index 99. Is it true?

```python
print(bigcolumn.index)
print(bigcolumn.index[-1] == 99)
```

### Step 7. Reindex the DataFrame so it goes from 0 to 299

```python
bigcolumn = bigcolumn.reset_index(drop=True)
bigcolumn.reindex(list(range(300)))
```
