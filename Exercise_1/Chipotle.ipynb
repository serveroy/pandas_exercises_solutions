### Step 9. Which was the most-ordered item? 

```python
# Creating a dictionary order_dict.
# key = item name (unique), value = number of orders for this item

# orders_dict = {}

# item_names = chipo.iloc[:]['item_name']
# unique_items = list(set(item_names))

# for item in unique_items:
#     orders_count = 0
#     for index in range(chipo.shape[0]):
#         if (item_names[index] == item):
#             orders_count += chipo.iloc[index]['quantity']
#     orders_dict[item] = orders_count

# orders_list = list(orders_dict.values())
# max_orders_index = orders_list.index(max(orders_list))

# most_ordered_item = unique_items[max_orders_index]
# print(most_ordered_item)


# A better solution: (idxmax gives the index for which the column is maximal)
grouping_by_quantity = chipo.groupby(["item_name"])[["quantity"]].sum()
most_ordered = grouping_by_quantity.idxmax().iloc[0]
most_ordered
```

### Step 10. For the most-ordered item, how many items were ordered?

```python
# most_orders = orders_dict[most_ordered_item]
# print(most_orders)

grouping_by_quantity.loc[most_ordered]["quantity"]
```

### Step 11. What was the most ordered item in the choice_description column?

```python
grouping_by_choice_description = chipo.groupby(["choice_description"])[["quantity"]].sum()
most_ordered_choice = grouping_by_choice_description.idxmax().iloc[0]
most_ordered_choice
```

### Step 14. How much was the revenue for the period in the dataset?

```python
new_prices = chipo['item_price']
quantity = chipo['quantity']
total_revenue = sum(new_prices * quantity)
print(total_revenue)
```

### Step 15. How many orders were made in the period?

```python
# orders_list = list(chipo.iloc[:]['order_id'])
# orders_count = orders_list[-1]
# print(orders_count)

orders_count = chipo.iloc[chipo.shape[0] - 1]['order_id']
orders_count
```

### Step 16. What is the average revenue amount per order?

```python
# Solution 1
total_revenue/orders_count
```

```python
# Solution 2
orders_count_second_way = len(list(set(list(chipo.iloc[:]['order_id']))))
total_revenue/orders_count_second_way
```
